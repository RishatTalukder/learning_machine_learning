# <a id='toc2_'></a>[Introduction to Decision Tree & Random Forest](#toc0_)

## <a id='toc2_1_'></a>[What is Decision Tree?](#toc0_)

`Decision Tree` is a `supervised learning` algorithm that can be used for both `classification` and `regression` problems. It creates a model that predicts the value of a target variable by learning simple decision rules inferred from the data features.

The `decision tree` algorithm builds the classification or regression models in the form of a `tree structure`. It breaks down a dataset into smaller and smaller `subset` while at the same time an associated `decision tree` is incrementally developed. The final result is a `tree` with `decision` nodes and `leaf` nodes. A `decision` node has two or more `branches` each representing values for the `attribute` tested. Leaf node represents a decision on the numerical target. The `topmost` decision node in a tree which corresponds to the best predictor called `root node`. Decision trees can handle both categorical and numerical data.
 

 Many things happends under the hood when we train a `decision tree` model and this is a great example of how `math and programming` can be used to solve real world problems. But we don't need to know all the math behind it to use it. I'll talk more about it later. 

## <a id='toc2_2_'></a>[What is Random Forest?](#toc0_)

`Random Forest` is a `supervised learning` is a `ensemble learning` algorithm. It's basically the `decision tree` algorithm with some extra steps. It creates a `forest` by training several `decision tree` on different `subsets` of the whole dataset. It uses `bagging` and `feature randomness` when building each individual tree to try to create an `un-correlated` forest of trees whose prediction by committee is more accurate than that of any individual tree.

## <a id='toc2_3_'></a>[Why Decision Tree & Random Forest?](#toc0_)

- Easy to understand and interpret
- Can handle both categorical and numerical data
- Requires less data preprocessing from the user
- Can be used for feature engineering
- Can be used for both classification and regression problems
- Can be used for multi-output problems
- Can be used for both binary and multi-class classification problems

## <a id='toc2_4_'></a>[Pros and Cons of Decision Tree & Random Forest](#toc0_)

### <a id='toc2_4_1_'></a>[Pros of Decision Tree](#toc0_)

- Decision trees are easy to interpret and visualize.
- It can easily capture Non-linear patterns.
- It requires fewer data preprocessing from the user, for example, there is no need to normalize columns.
- It can be used for feature engineering such as predicting missing values, suitable for variable selection.
- The decision tree has no assumptions about distribution because of the non-parametric nature of the algorithm. (Source)
- The algorithm is fast, scalable, and easy to use, it can easily capture Non-linear patterns.

### <a id='toc2_4_2_'></a>[Cons of Decision Tree](#toc0_)

- Sensitive to noisy data. It can overfit noisy data.
- The small variation(or variance) in data can result in the different decision tree. This can be reduced by bagging and boosting algorithms.
- Decision trees are biased with imbalance dataset, so it is recommended that balance out the dataset before creating the decision tree.
- Information gain in a decision tree with categorical variables gives a biased response for attributes with greater no. of categories.
- Calculations can become complex when there are many class labels.

### <a id='toc2_4_3_'></a>[Pros of Random Forest](#toc0_)

- It is one of the most accurate learning algorithms available. For many data sets, it produces a highly accurate classifier.
- It runs efficiently on large databases.
- It can handle thousands of input variables without variable deletion.
- It gives estimates of what variables are important in the classification.
- It generates an internal unbiased estimate of the generalization error as the forest building progresses.
- It has an effective method for estimating missing data and maintains accuracy when a large proportion of the data are missing.

### <a id='toc2_4_4_'></a>[Cons of Random Forest](#toc0_)

- Random forests have been observed to overfit for some datasets with noisy classification/regression tasks.
- For data including categorical variables with different number of levels, random forests are biased in favor of those attributes with more levels. Therefore, the variable importance scores from random forest are not reliable for this type of data.
- If the data contain groups of correlated features of similar relevance for the output, then smaller groups are favored over larger groups.
- Random Forest can feel like a black box approach for a statistical modelers we have very little control on what the model does. You can at best try different parameters and random seeds.

# <a id='toc3_'></a>[Decision Tree](#toc0_)

## <a id='toc3_1_'></a>[How Decision Tree Works?](#toc0_)

here's some `BIIIIIIIIG` theory for you.

Suppose we have data set like this:

| x | y | z | class |
| --- | --- | --- | --- |
| 1 | 1 | 1 | A |
| 1 | 1 | 0 | A |
| 0 | 0 | 1 | B |
| 1 | 0 | 0 | B |

Now, we want to predict the `class` of a new data point `x=1, y=0, z=1`. So, how do we do that?

SO, you can use other algorithm to solve this but as this is an example of `decision tree` so we will use `decision tree` to solve this problem.

SO, We make a tree.

### <a id='toc3_1_1_'></a>[Making a tree](#toc0_)

We have `x`, `y`, `z` as our `features` and `class` as our `target`. So, we make a tree that has perfect `split` on `x`, `y`, `z` and `class`.

So, we make splits based on `x`, `y`, `z` and `class` .

for `x`:

| x= 0 | x= 1 |
| --- | --- |
| - | A |
| - | A |
| B | - |
| - | B |



for `y`:

| y= 0 | y= 1 |
| --- | --- |
| - | A |
| - | A |
| B | - |
| B | - |

for `z`:

| z= 0 | z= 1 |
| --- | --- |
| - | A |
| A | - |
| - | B |
| B | - |

As you can see we don't have a perfect split for `x` and `z` but we have a perfect split for `y`. So, we will use `y` as our `root node`.

### <a id='toc3_1_2_'></a>[Root Node](#toc0_)

So, we have `y` as our `root node`. Now, we can make `decision` based on `y`. So, if `y=0` then we will go to the `left` and if `y=1` then we will go to the `right`.

We don't need to make splits on the other features because we already have a perfect split on `y`.

## <a id='toc3_2_'></a>[How do we make a perfect split on a feature?](#toc0_)

We make a perfect split on a feature by calculating the `entropy` of that feature.

### <a id='toc3_2_1_'></a>[What is Entropy?](#toc0_)

`Entropy` is a measure of `impurity` in a bunch of examples. The more `impure` a bunch of examples is, the more `entropy` it has.

### <a id='toc3_2_2_'></a>[How do we calculate Entropy?](#toc0_)

We calculate `entropy` by using this formula:

$$Entropy = -\sum_{i=1}^{n} p_i log_2(p_i)$$

where, $p_i$ is the probability of the $i^{th}$ class.

This is a porcess is very long and it's just finding the best `split` for a `feature`. 

Here's am example:

Suppose you have this table:

| Attribute 1 | Attribute 2 | Attribute 3 | Attribute 4 | Class |
|-------------|-------------|-------------|-------------|-------|
| A           | B           | C           | D           | 0     |
| B           | A           | D           | C           | 1     |
| C           | D           | A           | B           | 0     |
| D           | C           | B           | A           | 1     |
| A           | C           | B           | D           | 0     |
| B           | D           | A           | C           | 1     |
| C           | A           | D           | B           | 0     |
| D           | B           | C           | A           | 1     |

Now, we will find the entropy of the whole table. So, we will find the probability of `class 0` and `class 1`.

$$p_0 = \frac{4}{8} = 0.5$$

$$p_1 = \frac{4}{8} = 0.5$$

then we fond the `entropy` of the whole table.

$$Entropy = -p_0 log_2(p_0) - p_1 log_2(p_1)$$

$$Entropy = -0.5 log_2(0.5) - 0.5 log_2(0.5)$$

$$Entropy = -0.5 * -1 - 0.5 * -1$$

$$Entropy = 1$$

Now, Have to go through the same process for every feature. But this time we can see that every feature has 4 `unique` values. So, we have to find the `entropy` of every `unique` value of every feature.

So, we start with the `first` feature.

*** Attribute 1 ***

| Attribute 1 | Class |
|-------------|-------|
| A           | 0     |
| B           | 1     |
| C           | 0     |
| D           | 1     |
| A           | 0     |
| B           | 1     |
| C           | 0     |
| D           | 1     |

Now, we find he `entropy` for the `A` value of `Attribute 1`.

$$p_A0 = \frac{2}{2} = 1$$

$$p_A1 = \frac{0}{2} = 0$$

$$Entropy_A = -p_A0 log_2(p_A0) - p_A1 log_2(p_A1)$$

$$Entropy_A = -1 log_2(1) - 0 log_2(0)$$

$$Entropy_A = -1 * 0 - 0 * -1$$

$$Entropy_A = 0$$

Now we do the same for `B`, `C` and `D`.

$$Entropy_B = 0$$

$$Entropy_C = 0$$

$$Entropy_D = 0$$

Now, this will be the cruicial part. We find the information gain of `Attribute 1`.

$$Information Gain = Entropy - \sum_{i=1}^{n} \frac{p_i}{p} Entropy_i$$

$$Information Gain = 1 - \frac{2}{8} * 0 - \frac{2}{8} * 0 - \frac{2}{8} * 0 - \frac{2}{8} * 0$$

$$Information Gain = 1$$

Now, you do the rest of the features and find the `information gain` of every feature. 

The attribute with the highest `information gain` will be the `root node`. Meaning that we will make a `split` on that attribute.

The we will have to do the same process for the `left` and `right` side of the `root node`. And we will keep doing this until we have a perfect `split` on every `feature`.

Now, i told you before that the `decision tree` algorithm has math behind it. But this process is very long and repetitive. 

To learn more and see how this process is done, you can just search on `youtube` `decision tree entropy and information gain` and you will find a lot of videos on this topic.


# <a id='toc4_'></a>[Random Forest](#toc0_)

## <a id='toc4_1_'></a>[How Random Forest Works?](#toc0_)

Random forrest is multiple decision tree. So, it's basically the `decision tree` algorithm. Nothing new here.

But the general concept is that we make multiple `decision tree` and then we take the `average` of all the `decision tree` to make our final `decision tree`.

*** Why do we take the average of all the decision tree?

Because we want to reduce the `variance` of our model. If we take the `average` of all the `decision tree` then we can reduce the `variance` of our model.

*** How do we make multiple decision tree?

We make multiple `decision tree` by using `bagging` and `feature randomness`. 

The `decision trees` are made by taking `random` samples of the data and `random` features of the data. So, every time we make a `decision tree` we take `random` samples of the data which can lead to different `decision tree` with different `accuracy`. So, we take the `average` of all the `decision tree` to make our final `decision tree`. Which will have less `variance` than a single `decision tree`.

*** What is Bagging?

It's another algorithm similar to `random forest`. But it reduces the `variance` of our model by taking `random` samples of the data. It doesn't take `random` features of the data. So, it's not as good as `random forest`.

## <a id='toc5_2_'></a>[Decision Tree](#toc0_)

```python
from sklearn.model_selection import train_test_split
```

Now that we have imported the `train_test_split` class, we can split the data into `training` and `testing` data. As we are using the `Kyphosis` attribute as our `target` so we will drop it from the `X` variable. And we will use the `Kyphosis` attribute as our `y` variable.

```python
X = df.drop('Kyphosis', axis=1)
y = df['Kyphosis']
```

Now we spilt into `training` and `testing` data.

```python
# training and testing data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

We can now use the `DecisionTreeClassifier` class to train our model.

```python
from sklearn.tree import DecisionTreeClassifier

# decision tree classifier
dtree = DecisionTreeClassifier()
```

As we have the model in hand we can now fit the model with the `training` data.

```python
dtree.fit(X_train, y_train)
```

Our model is trained. Now we can make predictions with the `testing` data. Before testing the model I want to show you how we can just make a prediction with the model of a `single data point.`

 Let's see the first row of the `X_test` data.

```python
# the  first row of the test data
print(X_test.iloc[0])
```

The first row of the `X_test` data is `age = 159`(the data is made up as you can see), `number = 4` and `start = 13`. Now, we see the first row of the `y_test` data.

```python
y_test.iloc[0]
```

In this case we can say that when `age = 159`, `number = 4` and `start = 13` then the `Kyphosis` was `absent` after the operation. Let's see what our model predicts.

```python
# prediction of the first row of the test data

dtree.predict([X_test.iloc[0]])
```

> We need to pass a `list` of `lists` meaning a `2D array` to the `predict` method. Because the `predict` method expects a `2D array` Because when we pass the x_test dataframe it's a `2D array` SO, if we are getting the first item of the `x_test` dataframe then we need to pass a `2D array` with the first item of the `x_test` dataframe.

SO, the model predicts that when `age = 159`, `number = 4` and `start = 13` then the `Kyphosis` was `absent` after the operation. Which is correct.

Now we cannot do this with each and every data point. So, we will just use the `predict` method to predict the `y_pred` of the `X_test` data.

```python
#prediction of the whole test data

predictions = dtree.predict(X_test)
predictions[:5] # first five predictions
```

The whole testign prediction is done. Now we can see how accurate our model is.

```python
# evaluating the model
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, predictions))

print(confusion_matrix(y_test, predictions))
```

This is the `confusion matrix` of our model and we can see that the model shows `.84 or 84%` accuracy. Which is pretty good.

The precision in prediction `present` is `1.0` which means that when the model predicts that the `Kyphosis` was `present` after the operation `100%` of the time it was correct. I think that's pretty good.

## <a id='toc5_3_'></a>[Random Forest](#toc0_)

The process is almost the same as the `decision tree`. We just need to import the `RandomForestClassifier` class.

```python
# Random Forests

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100)
```

We just made the model object just like we did with the `decision tree` model. Now we can fit the model with the `training` data.

> n_estimators is the number of `decision tree` we want to make. The more `decision tree` we make the more `accurate` our model will be. But it will take more time to train the model. So, we need to find the `sweet spot` where we have the most `accurate` model with the least amount of time to train the model.

I use `100` as the number of `decision tree` to make. As we are working with a small dataset so it doesn't take much time to train the model. But if you are working with a large dataset then you can use `10` or `20` as the number of `decision tree` to make.

Now we can make predictions with the `testing` data.

SO, again we spilt into `training` and `testing` data. And we fit the model with the `training` data.

```python
from sklearn.model_selection import train_test_split

X = df.drop('Kyphosis', axis=1)
y = df['Kyphosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

Now, we fit to train the model. And we make predictions with the `testing` data.

```python
rfc.fit(X_train, y_train)

predictions = rfc.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, predictions))

print(confusion_matrix(y_test, predictions))
```

WEll, THAT WAS ***DISAPPOINTING***. The `decision tree` model was more `accurate` than the `random forest` model. 

This happened because the dataset is very small AND when we use `random forest` we take `random` samples of the data and `random` features of the data. So, every time we make a `decision tree` we take `random` samples of the data which can lead to different `decision tree` with different `accuracy`. 

As the dataset is very small so the `random forest` model takes even smaller `samples` of the data. So, the `decision tree` model was more `accurate` than the `random forest` model.

But most of the time the `random forest` model is more `accurate` than a single `decision tree` model. So, don't be disappointed.

# <a id='toc6_'></a>[Visualization](#toc0_)

One of the main goal of this article is to learn how to visualize the `decision tree` in python. It can be very useful when you are working with a large dataset and you want to visualize the `decision tree` to see how it works.

We have our data and we went though the process of training the model. Now we can visualize the `decision tree` model. Again a reminder of the process of training the model.

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix

df = pd.read_csv('datasets/kyphosis.csv')

X = df.drop('Kyphosis', axis=1)
y = df['Kyphosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

dtree = DecisionTreeClassifier()

dtree.fit(X_train, y_train)

predictions = dtree.predict(X_test)

print(classification_report(y_test, predictions))

print(confusion_matrix(y_test, predictions))
```

Well now it's even better then it was before. Let's visualize the `decision tree` model.

```python
from sklearn import tree
import matplotlib.pyplot as plt

# visualizing the decision tree
plt.figure(figsize=(14,14))

tree.plot_tree(
    dtree,
    feature_names=X.columns,
    class_names=y.unique(),
    filled=True,
    )

plt.show()
```

Sklearn has a `tree` module that we can use to visualize the `decision tree` model and it'll be very easy to do.

We pass our feature names to the `feature_names` parameter which is the `X.columns` and we can pass the `class_names` parameter which is the `y.unique()` which will return the unique values of the `y` variable.

You can see the `decision tree` model, it's `split` and `leaf` nodes & the `gini index` is also shown with other information. I'll leave the rest to you and use this method to visualize the `decision tree` model.
