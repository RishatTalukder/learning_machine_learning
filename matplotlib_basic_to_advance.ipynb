# what is pyplot?

- `pyplot` is a collection of command style functions that make matplotlib work like MATLAB. Each `pyplot` function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

To make a plot we need data. So, lets make some data using the `numpy` library.

```python
# a list of evenly spaced numbers from 0 to 5, with 11 numbers
x = np.linspace(0, 5, 11)
# we need another array
y = x ** 2
print(x)
print(y)
# y is exactly x squared
```

there are two ways to make a plot in matplotlib:

- `Functional method`
- `Object Oriented method`

# Functional Method

```python
plt.plot(x,y)
# plt.show() takes two arguments, x and y
# this will create a graph where x is the horizontal axis and y is the vertical axis

#if not using jupyter notebook, use plt.show() to show the plot
# plt.show()
```

```python
#we can add labels to the plot
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# multiplot
# plt.subplot(nrows, ncols, plot_number)

# plt.subplot(1,2,1) means 1 row, 2 columns, plot number 1
plt.subplot(1,2,1)
plt.plot(x,y,'r') # 'r' is the color red

# plt.subplot(1,2,2) means 1 row, 2 columns, plot number 2
plt.subplot(1,2,2)
plt.plot(y,x,'b') # 'b' is the color blue
```

# Object Oriented Method

idea is to create a `figure` object, then call methods off of it
this way we can have full control of our plot

## Add axes to the figure

> `add_axes([left, bottom, width, height])`
- left, bottom, width, height are all in percentage of the canvas
- left is the distance from the left edge of the canvas
- bottom is the distance from the bottom edge of the canvas
- width is the width of the canvas
- height is the height of the canvas

```python
# Object Oriented Method

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# now we have an empty canvas and we can plot on it
axes.plot(x,y)

# we can add labels to the plot
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# we can add more than one axes to the canvas
fig = plt.figure()
# the first axes
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x,y)
axes1.set_title('Larger Plot')

# the second axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(y,x)
axes2.set_title('Smaller Plot')

```

### subplots using object oriented method

```python
# subplots() method returns a tuple containing a figure and axes object(s)
fig,axes = plt.subplots() # this is a tuple unpacking
# in this case we only have one axes because we didn't specify the number of rows and columns
```

```python
# we can specify the number of rows and columns

fig,axes = plt.subplots(nrows=3, ncols=3)
```

```python
# empty canvas
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
# subplots() method takes two arguments, nrows and ncols (number of rows and number of columns)
# it returns a tuple containing a figure and axes object(s)
fig,axes = plt.subplots(nrows=1, ncols=2) # this is a tuple unpacking

# we can iterate through the axes, because it is a numpy array containing two axes
axes[0].plot(x,y)

axes[1].plot(y,x)
```

```python
#if you look at the type of axes, it is a numpy array
type(axes)
```

```python
#so we can iterate through the axes
fig,axes = plt.subplots(nrows=1, ncols=2) # this is a tuple unpacking

for current_ax in axes:
    current_ax.plot(x,y)
```

# Figure Size and DPI(dots per inch)

```python
fig = plt.figure(figsize=(3,2,),dpi=100) # figsize is a tuple that specifies the width and height of the figure in inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#we can experiment by changing the figsize and dpi
#fig = plt.figure(figsize=(8,2,),dpi=100) 

fig = plt.figure(figsize=(8,2,),dpi=50) # figsize is a tuple that specifies the width and height of the figure in inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# we cna also do the same thing with subplots
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2,),dpi=100) # figsize is a tuple that specifies the width and height of the figure in inches   

axes[0].plot(x,y)
axes[1].plot(y,x)

# if there is some overlapping between the two plots
plt.tight_layout()
```

# lagend

- what is a legend?

> A legend is an area describing the elements of the graph. In the matplotlib library, theres a function called legend() which is used to Place a legend on the axes.

simply `lagend()` helps clarify what is being plotted.

```python
# lets make a another plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Title')
```

```python
#to use the legend method, we need to add a label to the plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend(loc = 0) # loc=0 means let matplotlib decide the best location for the legend
```

# Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='X Squared',color="green",linewidth=4, alpha = 0.5, linestyle='--', marker='o', markersize=10, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='red')

# alpha is the transparency of the line
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
