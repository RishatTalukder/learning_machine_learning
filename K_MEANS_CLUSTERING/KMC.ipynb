# Introduction

`K-means` is a type of `unsupervised learning` of `unlabeled data`. It is one of the most popular `clustering` algorithms. Here a `cluster` refers to a collection of `data points` aggregated together because of certain similarities. You'll define a target number `k`, which refers to the number of centroids you need in the `dataset`. 

A `centroid` is the `imaginary` or `real` location representing the `center` of the `cluster/group`. Every `data point` is allocated to each of the `clusters` through reducing the `intra-cluster` distances and increasing the `inter-cluster` distances. In other words, the `K-means` algorithm identifies `k` number of `centroids`, and then allocates every `data point` to the nearest `cluster`, while keeping the `centroids` as small as possible. 

> Intra-cluster distance: It is the distance within the cluster. The smaller the better.
> Inter-cluster distance: It is the distance between the clusters. The larger the better.

# K-means Clustering

## What is Clustering?

`Clustering` is the task of dividing the `population` or `data points` into a number of `groups` such that `data points` in the same groups are more `similar` to other `data points` in the same group and dissimilar to the `data points` in other groups. It is basically a collection of objects on the basis of `similarity` and `dissimilarity` between them.


## What is K-means Clustering?

`K-means` is a `unsuperivsed` `learning` algorithm that is used to solve the `clustering` problems in `unlabeled or labeled` data. 

Basically, `K-means clustering` algorithm takes the `input` `data` and puts it into a group. Now how do we get the `group`? That's where the `K-means` algorithm comes in. The `K` in `K-means` refers to the number of `clusters` we want to divide our `data` into.

We give the `k` value to the `K-means` algorithm and it divides the `data` into `k` number of `clusters` or `groups`. Then the we get the `centroid` of each `cluster`. The `centroid` is the `center` of the `cluster`. 

The `K-means` algorithm then assigns each `data` point to the `closest` `centroid`. The `closest` `centroid` is the `centroid` that is the `least` distance away from the `data` point.

The `K-means` algorithm then calculates the `mean` of all the `data` points that are assigned to the `centroid`. This `mean` is the new `centroid` of the `cluster`. The `K-means` algorithm then repeats the process of assigning the `data` points to the `closest` `centroid` and calculating the `mean` of all the `data` points assigned to the `centroid` until the `centroids` don't change.

The `K-means` algorithm then stops and we get the `clusters` of the `data`.

SO, you can say that it's a `iterative` algorithm that divides the `data` into `k` number of `clusters` and assigns each `data` point to the `closest` `centroid` and then calculates the `mean` of all the `data` points assigned to the `centroid` and then repeats the process until the `centroids` don't change.

## How does K-means Clustering work?

The `K-means` algorithm works as follows:

1. First we initialize `k` centroids. We can initialize them in a few different ways. One way is to randomly select `k` data points from the `data` and use those as the `centroids`. Another way is to randomly generate `k` `centroids` and use those as the `centroids`.
2. Then we assign each `data` point to the `closest` `centroid`. The `closest` `centroid` is the `centroid` that is the `least` distance away from the `data` point.
3. Then we calculate the `mean` of all the `data` points that are assigned to the `centroid`. This `mean` is the new `centroid` of the `cluster`.
4. Then we repeat the process of assigning the `data` points to the `closest` `centroid` and calculating the `mean` of all the `data` points assigned to the `centroid` until the `centroids` don't change.
5. The `K-means` algorithm then stops and we get the `clusters` of the `data`.

This might help you understand the `K-means` algorithm better.




as you can see in that image, the `K-means` algorithm divides the `data` into `2` `clusters` and assigns each `data` point to the `closest` `centroid` and you should also see that the `centroids` change after each iteration until the `centroids` don't change anymore.

This is how the `K-means` algorithm works. 

## Example

Let's say we have a `data` of `9` `data` points and we want to divide the `data` into `2` `clusters`.

Let's say our data points are: 2, 4, 10, 12, 3, 20, 30, 11, 25

1. **Initialization**: First, we need to initialize the centroids. Let's choose 2 random points from our data as the initial centroids. Let's say we choose 2 and 20.

2. **Assignment**: Next, we assign each data point to the nearest centroid. This gives us the following clusters:

   - Cluster 1 (centroid = 2): 2, 3, 4
   - Cluster 2 (centroid = 20): 10, 12, 20, 30, 11, 25

3. **Update**: Now, we update the centroids by calculating the mean of each cluster:

   - New centroid of Cluster 1 = mean(2, 3, 4) = 3
   - New centroid of Cluster 2 = mean(10, 12, 20, 30, 11, 25) = 18

4. **Repeat**: We repeat the assignment and update steps until the centroids do not change anymore. In this case, the centroids do not change after the first iteration, so we are done.

So, the final clusters are:

- Cluster 1: 2, 3, 4
- Cluster 2: 10, 12, 20, 30, 11, 25

This is a simplified example of how the K-means algorithm works. In practice, the data would be multi-dimensional, and the Euclidean distance would be used to determine the nearest centroid.



# Implementing K-means Clustering using Python and Scikit-learn

Now that you know what `K-means` is and how it works, let's implement `K-means` using `python` and `scikit-learn` library.

We will use built-in `blobs` dataset from `scikit-learn` library. The `blobs` dataset is a `synthetic` dataset that is useful for `clustering` algorithms. It is a great way to test `clustering` algorithms because the `clusters` are well defined and the `data` is easy to visualize.

SO, let's start by importing the necessary libraries.



```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs #import make_blobs dataset from sklearn

# ignore warnings from the libraries
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

Now that we have imported the necessary libraries, We can now load the `blobs` dataset from `scikit-learn` library and then we can visualize the `data` using a `scatter plot`.

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

Let me explain what I just did in the code above.

1. I loaded the `blobs` dataset from `scikit-learn` library using the `make_blobs` function. 
2. I then made a variable `data` and called the `make_blobs` function and passed in the `n_samples` parameter as `200` which means that I want `200` `data` points in the `data`. 
3. I also passed in the `n_features` parameter as `2` which means that I want `2` `features` in the `data` and I also passed in the `centers` parameter as `4` which means that I want `4` `clusters` in the `data`.
4. `cluster_std` is the `standard deviation` of the `clusters`. I passed in `1.8` as the `cluster_std` parameter. (this is not necessary but I did it to make the `clusters` more spread out)
5. random_state is the `seed` used by the `random number generator`. I passed in `101` as the `random_state` parameter.

let's see the data.

```python
data
```

Well, that's a problem. The data has `2` `features` represented as a `2D` `array` and another `array` that represents the `cluster` of each `data` point. SO, if we want to visualize the `featued` data, we can just try to print the first index of the `data` and see what it looks like.

```python
data[0].shape
```

feature data shape. We can see it has `2` `features` and `200` `data` points. Just like we wanted. Now let's store the feature data in a variable called `features`.

```python
features = data[0]
cluster = data[1]
print(features[:10])
```

Here is the first 10 feature data points. Now here is a point I would like to remind you.

> DO you remember I said that the `K-means` algorithm is an `unsupervised` `learning` algorithm for `unlabeled` `data`? 

That's why we are using an Unlabeled dataset. 

Now let's visuliaze the `data` using a `scatter plot`.

```python

plt.scatter(features[:,0], features[:,1], c=cluster, cmap='viridis')
```

If this goes over your head.. Then please recap the `numpy` 2D array and `matplotlib` library from my previous articles. Let me break it down for you.

> features[:,0] means that I want all the rows(:) of the first column(0) of the `features` array.
> features[:,1] means that I want all the rows(:) of the second column(1) of the `features` array.
> c=cluster means that I want to color the `data` points based on the `cluster` of each `data` point. The `cluster` is the third `array` in the `data` that represents the `cluster` of each `data` point.
> cmap='rainbow' means that I want to use the `rainbow` color map to color the `data` points.

Then we see that the `data` is divided into `4` `clusters` and each `cluster` is colored differently. Like we wanted it to be.

Now we can implement the `K-means` algorithm using `python` and `scikit-learn` library.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4) # 4 because we know that there are 4 clusters
```

Now, we fit the data to the `K-means` algorithm.

```python
kmeans.fit(features)
```

The model is trained and we can now get the `cluster` centers and the `labels` of the `data`.

```python
clusters = kmeans.cluster_centers_
print(clusters)

labels = kmeans.labels_
print(labels)
```

Now, we visualize if we have correctly clustered the data like we did before.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(features[:,0], features[:,1], c=labels, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(features[:,0], features[:,1], c=cluster, cmap='rainbow')
```

> the colores don't mean anything. They are just for visualization purposes.

SO, as you can see the clusters are almost the same as before. The `K-means` algorithm has done a good job of clustering the `data`.

Now As this is an `unsupervised` `learning` algorithm, we don't have the `labels` of the `data` and the target clusters are not known. SO, we can not calculate the `accuracy` of the `model`.
